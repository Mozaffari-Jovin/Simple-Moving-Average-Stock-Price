```python
# Simple moving average using Numpy (finding the sum of elements in the window and dividing it with window size)

window_size = 6 # consider a window of size 6
i = 0 
moving_averages = [] # initialize an empty list storing moving averages
  
while i < np.size(values_cl) - window_size + 1:
    window_average = round(np.sum(values_cl[i:i+window_size]) / window_size, 2) # calculate the average of the current window
    moving_averages.append(window_average) # store the average of the current window in the moving average list
    i += 1 # shift the window to right by one position

print(moving_averages)
```

```python
plt.plot(values_cl[:105], label="Price")
plt.plot(moving_averages[:105], label="Average")
plt.title("Time history of the stock's close price vursus its 6-day simple moving average (Numpy)")
plt.xlabel("Working day")
plt.ylabel("Value") 
plt.legend()
plt.show() 
```

```python
# Simple moving average using Pandas (finding the sum of elements in the window and dividing it with window size)

numbers_series = pd.Series(values_cl) # convert array of integers to pandas series
windows = numbers_series.rolling(window_size) # specify window size the for observations
moving_averages = windows.mean() # calculate moving average for each window
moving_averages_list = moving_averages.tolist() # convert pandas series back to list
moving_averages_list_f = moving_averages_list[window_size - 1:] # discard null entries from the list

print(moving_averages_list_f)
```

```python
plt.plot(values_cl[:105], label="Price")
plt.plot(moving_averages_list_f[:105], label="Average")
plt.title("Time history of the stock's close price vursus its 6-day simple moving average (Pandas")
plt.xlabel("Working day")
plt.ylabel("Value") 
plt.legend()
plt.show() 
```
